```python
import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt

import IPython.display as display
```

```python
time_serires_data = pd.read_excel('../data/Learning_data_part1.xlsx')
```

```python
test_series = time_serires_data.iloc[:, 1].values
```

```python
# plot series
fig, ax = plt.subplots(figsize=(14, 6))

for indx in [1]:
    ax.plot(time_serires_data.iloc[:, indx].values, label=f'series {indx}')

ax.legend()
ax.grid(True)
ax.set_xlabel('$t$');
```

```python
# transfrom series shape
test_series = test_series.reshape((-1, 1))
```

```python
# pca model
pca_model = ProbalisitcPcaModel(test_series)

window_size = 50

discord_scores = pca_model.find_discord(window_size)
```

```python
# plot discord function
fig, ax = plt.subplots(figsize=(10, 6))

for indx in [1]:
    ax.plot(np.arange(window_size - 1, test_series.shape[0] - window_size + 1), discord_scores)

ax.grid(True)
ax.set_xlabel('window location');
ax.set_ylabel('Discord');
ax.set_title('PCA model');
```

```python
print(np.mean(discord_scores), np.std(discord_scores, ddof=1))
```
